```python
import yfinance as yf
from datetime import  datetime
import pandas as pd
```

```python
symbol = input("Symbol = ")
start_date = input(f"Start date = (YYYY-MM-DD)").strip()
end_date = input(f"End date = (YYYY-MM-DD)").strip()
interval = input("interval = (1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo)")  
mapping_symbol = {
    "0P0000HY8X.VN": "VNINDEX",
    "KO": "coca.cola",
    "BTC-USD": "btc",
    "ETH-USD": "eth"
}
stock = yf.Ticker(symbol)
```

```python
stock.info
```

```python
hist = stock.history()
```

```python
hist
```

```python
hist = stock.history(
    start=start_date,
    interval=interval,
    end=end_date
)
```

```python
str(hist.index.min().date())
```

```python
str(hist.index.max().date())
```

```python
hist.index = hist.index.strftime('%Y-%m-%d %H:%M:%S')
hist.index = pd.to_datetime(hist.index).tz_localize('UTC')
```

```python
hist
```

```python
symbol =  mapping_symbol.get(symbol, symbol)
```

```python
symbol = symbol.replace(".", "_").lower()
```

```python
file_name = f"../../data/{symbol}/{symbol}_to_now_{interval.lower()}.csv"
```

```python
file_name = file_name.replace("-", "_")
```

```python
file_name
```

```python

hist.to_csv(file_name)
```
